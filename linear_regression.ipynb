# 1- Simple Linear Regression

## Step-1 import Dataset

```python
import pandas as pd
df = pd.read_csv('ml_data_salery.csv')
df.head()
```

## Step-2 Splitting Dataset into Training data and Testing Data

```python
x = df[['YearsExperience']]
y = df[['Salary']]
```

```python
x.head()
```

```python
y.head()
```

## Step 3 Splitting Dataset into training and testing data (80/20)

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

## Step-3 Fit Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x_train, y_train)
model
```

## Step 4 Plotting of Trained Data

```python
import matplotlib.pyplot as plt
plt.scatter(x_train, y_train)
plt.plot(x_train, model.predict(x_train), color = 'red')
plt.xlabel("Years Experience")
plt.ylabel('Salary')
plt.title("Train Data")
plt.show()
```

## Step 5 Plotting of test data

```python
import matplotlib.pyplot as plt
plt.scatter(x_test, y_test)
plt.plot(x_test, model.predict(x_test), color = 'red')
plt.xlabel("Years Experience")
plt.ylabel("Salary")
plt.title("Test Data")
plt.show()
```

## Step 6 Checking/Evaluating our model 

```python
# Model Fitness/score
print(f"Score for training model = {model.score(x_train, y_train)*100}'%'")
print(f"Score for testing model = {model.score(x_test, y_test)*100}'%'")
```

## Step 7 Prediction For unknown Values

```python
# If we have 5 years of experience
model.predict([[5]])
```

```python
# predicting more values
salary = model.predict([[5],[4],[8],[3],[6]])
salary
```

```python
# predicting list of years experience 
x = [[5],[6],[9],[8],[7],[2]]
model.predict(x)
```

## Step 8 Checking model accuracy 

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

# Example where x has 30 samples and 6 features, and y has 30 target values
x = np.random.rand(30, 6)  # 30 samples and 6 features
y = np.random.rand(30)  
# Splitting dataset into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# creating and fitting model
model = LinearRegression().fit(x_train, y_train)

# predicted values
predicted_values = model.predict(x_test)
predicted_values

mse = mean_squared_error(y_test, predicted_values)
r2 = r2_score(y_test, predicted_values)
print(f"Mean Squared Error: {mse*100}")
print(f"R-squared: {r2*100}")
```

## Step 9 Saving our model 

```python
import joblib
joblib.dump(model, 'linear_salery.dot')
```
